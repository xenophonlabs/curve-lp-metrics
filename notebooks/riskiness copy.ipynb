```python
def compute_lagged_corr(series_a, series_b, lags):
    """
    Compute the lagged Pearson correlation between two series.
    
    Parameters:
    series_a (np.array): first time series
    series_b (np.array): second time series
    lags (np.array): array of lags at which to compute the correlation

    Returns:
    correlations (np.array): array of correlation coefficients at the specified lags
    """
    # Initialize the array to store the correlation coefficients
    correlations = np.zeros(len(lags))

    # Loop over the lags
    for i, lag in enumerate(lags):
        # Shift one of the series to compute the lagged correlation
        if lag < 0:
            corr, _ = pearsonr(series_a[:lag], series_b[-lag:])
        elif lag > 0:
            corr, _ = pearsonr(series_a[lag:], series_b[:-lag])
        else:
            corr, _ = pearsonr(series_a, series_b)
        
        # Store the correlation coefficient
        correlations[i] = corr

    return correlations
```

```python
# Create some example data
series_a = np.random.normal(size=100)
series_b = series_a
lags = list(range(-10, 11))

# Compute the cross-correlation
cross_corr = compute_lagged_corr(series_a, series_b, lags)

f, ax = plt.subplots()

ax.plot(lags, cross_corr)
```
